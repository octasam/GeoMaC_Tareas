```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def mesh(a, b, Nt):
    """
    Funcion para definir la malla
    """
    ht = (b-a) / Nt
    return ht

def exactSolution(t ,y0 ,lam):
    """
    Funcion para calcular la solucion analitica para la equacion logistica
    """
    return y0 / ( y0 + (1 - y0) * np.exp(-lam*t))

def forwardEuler (y , ht , lam ):
    """
    Resolviendo mediante el metodo de Euler hacia adelante
    """
    A = 1 - ht*lam
    An = [A]
    for i, val in enumerate (y[0:-1]):
        y[i+1] = lam*y[i]*(1-y[i])
        An.append(An[i] * A)
    return An


"""
Parametros iniciales del problema
"""

Nt = 4
Tmax = 10
ht = mesh(0,Tmax,Nt)
y0 = 0.01
lam = 10

t = np.linspace(0 ,Tmax, Nt+1)
yf = np.zeros(Nt+1)

yf[0] = y0

An = forwardEuler(yf, ht, lam)

"""
Calculo del error
"""

error_f = forwardEuler(yf, ht, lam)

tl = np.linspace(0 , Tmax , 100)

y_exacta = exactSolution(tl, y0, lam)
y_exac_p = exactSolution(t, y0, lam)

norma_error_f = np.linalg.norm(yf - y_exac_p,2)


Ecuacion = '$y(t) = y_0 / y_0 +(1-y_0)*e^{\lambda t}$,' + '$N_t$' + '= {} '.format(Nt) + ', $h_t$' + '= {:03.2f}'.format(ht)
Error = ', Error : FE = {:10.9f},'.format(norma_error_f)

"""
Implementacion de la graficacion de los resultados obtenidos
para la ecuacion logistica
"""

plt.style.use(['Solarize_Light2'])
plt.suptitle('Funcion Logistica', fontsize =14)
plt.plot(tl, y_exacta, 'g-', lw =3, label = 'Sol.Exacta')
plt.plot(t, yf, 'C7o--', label = 'Nt=4')
plt.title(Ecuacion + Error, fontsize =12 , color='blue')
plt.xlim(-0.1,1.0)
plt.ylim(-0.1,1.0)
plt.xlabel('$t$')
plt.ylabel('$y(t)$')
plt.legend(loc ='upper left', ncol=1 , framealpha=0.75, fancybox=True ,fontsize =10)
plt.grid(color ='w')
```
